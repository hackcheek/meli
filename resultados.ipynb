```python
import pandas as pd

df = pd.read_csv('datos/results_a.csv')
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df['desc_words'] = df['description'].str.split().apply(len)
df['hydrated_words'] = df['hydrated_description'].str.split().apply(len)
```

```python
df[['desc_words', 'hydrated_words']].plot()
```

### First

```python
first = df.sort_values('desc_words').iloc[0]
```

```python
print(first['description'])
print('-'*100)
print(first['hydrated_description'])
```
